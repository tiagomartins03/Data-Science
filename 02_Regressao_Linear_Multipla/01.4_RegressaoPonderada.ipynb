# Regressão Ponderada
---

A regressão ponderada é usada por estatísticos para diversos propósitos. Em especial, é importante para análise de pesquisas complexas. Os cientistas de dados podem achar a regressão ponderada útil em dois casos:

- Na ponderação de uma variância inversa, quando diferentes observações foram medidas com precisão diferente.
- Na análise de dados de forma agregada, de modo que a variável ponderada mostra quantas observações originais cada linha nos dados agregados representa.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Dados de exemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
weights = np.array([1, 1, 1, 2, 2])  # Pesos aplicados aos pontos

# Adicionar uma constante para o termo de interceptação
x = sm.add_constant(x)

# Realizar a regressão ponderada
model = sm.WLS(y, x, weights=weights)
results = model.fit()

# Exibir os resultados da regressão
print(results.summary())

# Fazer previsões para os valores ajustados
y_pred = results.predict(x)

# Visualizar os dados e a regressão ajustada
plt.scatter(x[:, 1], y, label='Dados', color='blue')
plt.plot(x[:, 1], y_pred, label='Regressão Ponderada', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
