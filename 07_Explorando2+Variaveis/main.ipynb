# Tabelas de contigência

```python
# Um registro das contagens entre duas ou mais variáveis categóricas
```

```python
import pandas as pd

# Criando um DataFrame fictício
data = {
    'Genero': ['Masculino', 'Feminino', 'Feminino', 'Masculino', 'Masculino', 'Feminino', 'Feminino', 'Masculino', 'Feminino', 'Masculino'],
    'Compra': ['Sim', 'Não', 'Sim', 'Não', 'Sim', 'Sim', 'Não', 'Sim', 'Não', 'Sim']
}

df = pd.DataFrame(data)

# Criando a tabela de contingência
tabela_contingencia = pd.crosstab(df['Genero'], df['Compra'], margins=True)

# Exibindo a tabela de contingência
print(tabela_contingencia)
```

# Compartimentação hexagonal

```python
# O gráfico de duas ou mais variáveis métricas com os registros compartimentados em hexágonos
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregar o conjunto de dados Iris
df = sns.load_dataset('iris')

# Criar um gráfico de compartimentação hexagonal
plt.figure(figsize=(10, 6))
plt.hexbin(df['sepal_width'], df['sepal_length'], gridsize=30, cmap='Blues')
plt.colorbar(label='Contagem de pontos')
plt.title('Gráfico de Compartimentação Hexagonal - Largura x Comprimento da Sépala')
plt.xlabel('Largura da Sépala (cm)')
plt.ylabel('Comprimento da Sépala (cm)')
plt.show()
```

# Gráfico de contorno

```python
# Um gráfico mostrando a densidade de duas variáveis numéricas como um mapa topográfico
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import kde

# Carregar o conjunto de dados Iris
df = sns.load_dataset('iris')

# Criar uma grade regular de pontos
x = df['sepal_width']
y = df['sepal_length']

# Estimar a densidade de pontos usando KDE (Kernel Density Estimation)
kde_estimator = kde.gaussian_kde([x, y])
xi, yi = np.linspace(x.min() - 0.5, x.max() + 0.5, 100), np.linspace(y.min() - 0.5, y.max() + 0.5, 100)
Xi, Yi = np.meshgrid(xi, yi)
Zi = kde_estimator(np.vstack([Xi.ravel(), Yi.ravel()]))
Zi = Zi.reshape(Xi.shape)

# Criar o gráfico de contorno
plt.figure(figsize=(10, 6))
contour = plt.contourf(Xi, Yi, Zi, levels=20, cmap='viridis')
plt.colorbar(contour, label='Densidade Estimada')
plt.title('Gráfico de Contorno - Largura x Comprimento da Sépala')
plt.xlabel('Largura da Sépala (cm)')
plt.ylabel('Comprimento da Sépala (cm)')
plt.show()
```

# Gráfico de violino

```python
# Semenlhante a um boxplot mas mostra a densidade
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o conjunto de dados Iris
df = sns.load_dataset('iris')

# Criar um gráfico de violino
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_width', data=df, palette='muted')
plt.title('Gráfico de Violino - Largura da Sépala por Espécie')
plt.xlabel('Espécie')
plt.ylabel('Largura da Sépala (cm)')
plt.show()
```
