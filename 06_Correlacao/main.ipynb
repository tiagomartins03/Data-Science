# Correlacao

```python
#  na estatística é uma medida que indica o grau de relacionamento entre duas variáveis. 
# Em outras palavras, ela nos mostra se duas variáveis tendem a variar juntas de forma consistente.
# Os valores variam de -1 até 1, sendo que quanto mais próximo de 1 maior a força entre as variáveis
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando um DataFrame com dados fictícios
data = {
    'idade': [23, 45, 31, 35, 65, 22, 34, 42, 53, 28],
    'salario': [50000, 100000, 75000, 85000, 120000, 48000, 69000, 94000, 105000, 52000],
    'experiencia': [2, 20, 10, 12, 30, 1, 8, 18, 25, 3],
    'satisfacao': [7, 9, 6, 8, 9, 6, 7, 8, 9, 6]
}

df = pd.DataFrame(data)

# Calculando a matriz de correlação
correlation_matrix = df.corr()

# Plotando a matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlação')
plt.show()
```

# Dispersão

```python
# O gráfico de dispersão é outra forma de investigar a correlação entre as variáveis. Quanto mais juntas estiverem as variáveis maior será a correlação entre elas
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando um DataFrame com dados fictícios
data = {
    'idade': [23, 45, 31, 35, 65, 22, 34, 42, 53, 28],
    'salario': [50000, 100000, 75000, 85000, 120000, 48000, 69000, 94000, 105000, 52000],
    'experiencia': [2, 20, 10, 12, 30, 1, 8, 18, 25, 3],
    'satisfacao': [7, 9, 6, 8, 9, 6, 7, 8, 9, 6]
}

df = pd.DataFrame(data)

# Criando gráficos de dispersão
plt.figure(figsize=(15, 10))

# Gráfico de dispersão de idade vs. salário
plt.subplot(2, 2, 1)
sns.scatterplot(data=df, x='idade', y='salario')
plt.title('Idade vs. Salário')

# Gráfico de dispersão de idade vs. experiência
plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='idade', y='experiencia')
plt.title('Idade vs. Experiência')

# Gráfico de dispersão de idade vs. satisfação
plt.subplot(2, 2, 3)
sns.scatterplot(data=df, x='idade', y='satisfacao')
plt.title('Idade vs. Satisfação')

# Gráfico de dispersão de salário vs. experiência
plt.subplot(2, 2, 4)
sns.scatterplot(data=df, x='salario', y='experiencia')
plt.title('Salário vs. Experiência')

plt.tight_layout()
plt.show()
```
