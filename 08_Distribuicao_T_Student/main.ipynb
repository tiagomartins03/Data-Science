# Distribuição

A distribuição t-Student, também conhecida simplesmente como distribuição t, é uma distribuição de probabilidade que surge frequentemente na estatística. Ela é utilizada quando se deseja fazer inferências sobre uma média populacional, mas o desvio padrão populacional é desconhecido.

## Parâmetros da Distribuição T

Ao contrário da distribuição normal, que requer dois parâmetros (média e desvio padrão), a distribuição t de Student requer apenas um parâmetro: Graus de liberdade.

A distribuição t-Student requer apenas um parâmetro, os graus de liberdade.
Os graus de liberdade correspondem ao tamanho da amostra menos um (n – 1).
Quanto maior o número de graus de liberdade, mais a distribuição t se assemelha à distribuição normal.

## Aplicações

A distribuição t-Student é amplamente utilizada para testes de hipóteses, permitindo testar hipóteses sobre uma média populacional quando o desvio padrão populacional é desconhecido.

A distribuição t-Student é utilizada em testes de hipóteses para médias populacionais com desvio padrão desconhecido.
Permite testar hipóteses sobre uma média populacional quando o desvio padrão populacional é desconhecido.

## Construção dos intervalos de confiança

A construção de intervalos de confiança é uma ferramenta estatística essencial para estimar a média populacional quando o desvio padrão populacional é desconhecido. Esse método é comumente utilizado para estabelecer a faixa de valores dentro da qual a média populacional provavelmente se encontra, com base em uma amostra. Aqui estão alguns pontos importantes sobre a construção de intervalos de confiança:

Permite estimar a média populacional quando o desvio padrão populacional é desconhecido
É útil para estabelecer a faixa de valores dentro da qual a média populacional provavelmente se encontra
Ajuda a compreender a incerteza associada à estimativa da média populacional

## Relação entre a Distribuição Normal

A distribuição t e a distribuição normal estão intimamente relacionadas, especialmente em amostras grandes. Compreender essa relação é fundamental para garantir a validade das inferências estatísticas. Aqui estão alguns pontos relevantes sobre a relação entre a distribuição t e a distribuição normal:

- Em amostras grandes, a distribuição t se aproxima da distribuição normal
- A distribuição t converge para a distribuição normal padrão quando o número de graus de liberdade tende ao infinito
- Em amostras grandes onde se pode assumir normalidade dos dados, a distribuição t e a distribuição normal conduzem aos mesmos resultados nos testes de hipóteses

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Gerando uma amostra aleatória de pesos (em kg) de uma população
np.random.seed(42)
sample_data = np.random.normal(loc=70, scale=10, size=1000)  # Média 70kg, desvio padrão 10kg, amostra de 15

# Calculando a média da amostra
sample_mean = np.mean(sample_data)

# Calculando o desvio padrão da amostra
sample_std = np.std(sample_data, ddof=1)

# Definindo o nível de confiança (95%)
confidence_level = 0.95
degrees_of_freedom = len(sample_data) - 1


# Calculando o intervalo de confiança usando a distribuição T-Student
confidence_interval = stats.t.interval(
    confidence_level,
    degrees_of_freedom,
    loc=sample_mean,
    scale=sample_std/np.sqrt(len(sample_data))
)

# Exibindo o intervalo de confiança
print(f'Intervalo de Confiança para a Média ({confidence_level*100}%): {confidence_interval}')

# Plotando a distribuição T-Student
x = np.linspace(-4, 4, 1000)
t_distribution = stats.t.pdf(x, df=degrees_of_freedom)

plt.figure(figsize=(10, 6))
plt.plot(x, t_distribution, label=f'T-Distribution (df={degrees_of_freedom})')

# Marcando o intervalo de confiança no gráfico
t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)
plt.axvline(-t_critical, color='red', linestyle='--', label='Critical Value')
plt.axvline(t_critical, color='red', linestyle='--')

# Personalizando o gráfico
plt.title('Distribuição T-Student com Intervalo de Confiança')
plt.xlabel('Valores de t')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)

# Exibindo o gráfico
plt.show()

```
