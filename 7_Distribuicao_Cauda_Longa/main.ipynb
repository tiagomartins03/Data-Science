# Distribuição Cauda Longa

Distribuições de cauda longa são aquelas em que a frequência de ocorrência dos valores mais comuns é muito maior do que a ocorrência de valores extremos, e é caracterizada por haver um grande número de valores extremos. Como exemplos de distribuições, temos a Distribuição de Pareto, Lei de Zipf e Leis de Potência. Alguns exemplos práticos são:

- Outliers têm um papel importante, muitas vezes não faz sentido serem descartados como anomalias — pelo contrário, podem ser justamente o que se deseja.
- O conceito de desvio padrão não faz sentido da mesma forma que faz sentido para uma distribuição normal.
- Médias e medianas passam a ser bastante diferentes (e significar coisas bastante diferentes entre si).

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulação de dados de vendas de produtos
np.random.seed(42)
num_products = 1000  # Número de produtos
sales = np.random.zipf(a=2.0, size=num_products)  # Distribuição Zipfian para simular a cauda longa

# Criar um DataFrame para organizar os dados
df = pd.DataFrame({
    'Produto': [f'Produto {i+1}' for i in range(num_products)],
    'Vendas': sales
})

# Ordenar os produtos por vendas (maior para menor)
df = df.sort_values(by='Vendas', ascending=False)

# Plotar o gráfico da distribuição de cauda longa
plt.figure(figsize=(10, 6))
plt.plot(df['Produto'], df['Vendas'], marker='o', linestyle='-')
plt.yscale('log')  # Escala logarítmica para visualizar a cauda longa
plt.xticks([], [])  # Remover os nomes dos produtos do eixo x para melhor visualização
plt.title('Gráfico de Cauda Longa das Vendas de Produtos')
plt.xlabel('Produtos (ordenados por vendas)')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()
```
