# Distribuição Exponencial

Analisa inversamente o experimento de Poisson

- Poisson: Análise de falhas por um intervalo
- Exponencial: Análise de intervalo por falha

A distribuição exponencial é freqüentemente usada em estudos de confiabilidade como sendo o modelo para o tempo até a falha de um equipamento muito utilizado para componentes eletrônicos

A distribuição de probabilidade do intervalo t entre **dois sucessos** consecutivos que ocorram dentro do intervalo de Poisson

Mostra a frequência

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo a taxa de chamadas por minuto (lambda)
lambda_rate = 5  # significa 5 chamadas por minuto

# Gerando um conjunto de dados usando a distribuição exponencial
# O tempo de espera entre chamadas é distribuído exponencialmente
data = np.random.exponential(1/lambda_rate, 1000)

# Plotando o histograma do tempo de espera entre chamadas
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True, color="green")

# Adicionando detalhes ao gráfico
plt.title("Distribuição Exponencial - Tempo de Espera entre Chamadas", fontsize=15)
plt.xlabel("Tempo de Espera (minutos)", fontsize=12)
plt.ylabel("Frequência", fontsize=12)

# Mostrando o gráfico
plt.show()
```

Mostra a probabilidade

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Definindo a taxa de chamadas por minuto (lambda)
lambda_rate = 5  # significa 5 chamadas por minuto

# Gerando um conjunto de dados usando a distribuição exponencial
data = np.random.exponential(1/lambda_rate, 1000)

# Criando os bins para o histograma
bins = np.linspace(0, np.max(data), 50)

# Calculando a probabilidade para cada bin usando a PDF da distribuição exponencial
pdf_values = expon.pdf(bins, scale=1/lambda_rate)

# Normalizando para que a soma das probabilidades seja igual a 1
bin_width = bins[1] - bins[0]
pdf_values_normalized = pdf_values * bin_width

# Plotando o histograma com a função densidade de probabilidade
plt.figure(figsize=(10, 6))
plt.hist(data, bins=bins, density=True, alpha=0.6, color='green', label='Histograma')
plt.plot(bins, pdf_values_normalized, 'r-', lw=2, label='PDF da Exponencial')

# Adicionando detalhes ao gráfico
plt.title("Distribuição Exponencial - Tempo de Espera entre Chamadas", fontsize=15)
plt.xlabel("Tempo de Espera (minutos)", fontsize=12)
plt.ylabel("Probabilidade", fontsize=12)
plt.legend()

# Mostrando o gráfico
plt.show()
```
