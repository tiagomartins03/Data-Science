# Amostragem aleatória

```python
# Elementos aleatoriamente obtidos para uma amostra
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados aleatórios
np.random.seed(42)
idade = np.random.randint(20, 60, 100)
renda = idade * np.random.randint(200, 500) + np.random.normal(0, 10000, 100)

# Amostragem aleatória
amostra_indices = np.random.choice(range(100), size=30, replace=False)
idade_amostra = idade[amostra_indices]
renda_amostra = renda[amostra_indices]

# Plotar dados originais
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(idade, renda, alpha=0.5, color='blue')
plt.title('Dados Originais')
plt.xlabel('Idade')
plt.ylabel('Renda')

# Plotar dados da amostra aleatória
plt.subplot(1, 2, 2)
plt.scatter(idade_amostra, renda_amostra, alpha=0.5, color='green')
plt.title('Amostra Aleatória')
plt.xlabel('Idade')
plt.ylabel('Renda')

plt.tight_layout()
plt.show()
```

# Amostragem estratificada

```python
# Divide a população em estratos e faz a amostragem alaeatórias em cada estado
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Gerar dados aleatórios
np.random.seed(42)
idade = np.random.randint(20, 50, 100)
renda = idade * np.random.randint(200, 500) + np.random.normal(0, 10000, 100)

# Criar faixas etárias
faixa_etaria = pd.cut(idade, bins=[19, 29, 39, 49], labels=['20-29', '30-39', '40-49'])

# Criar DataFrame
dados = pd.DataFrame({'idade': idade, 'renda': renda, 'faixa_etaria': faixa_etaria})

# Amostragem estratificada
treino, amostra_estratificada = train_test_split(dados, test_size=0.3, stratify=dados['faixa_etaria'], random_state=42)

# Plotar dados originais
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(dados['idade'], dados['renda'], alpha=0.5, color='blue')
plt.title('Dados Originais')
plt.xlabel('Idade')
plt.ylabel('Renda')

# Plotar dados da amostra estratificada
plt.subplot(1, 2, 2)
plt.scatter(amostra_estratificada['idade'], amostra_estratificada['renda'], alpha=0.5, color='green')
plt.title('Amostra Estratificada')
plt.xlabel('Idade')
plt.ylabel('Renda')

plt.tight_layout()
plt.show()
```

# Amostra aleatória simples

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gerar dados aleatórios
np.random.seed(42)
idade = np.random.randint(20, 60, 100)
renda = idade * np.random.randint(200, 500) + np.random.normal(0, 10000, 100)

# Criar DataFrame
dados = pd.DataFrame({'idade': idade, 'renda': renda})

# Amostragem aleatória simples
tamanho_amostra = 30
amostra_aleatoria_simples = dados.sample(n=tamanho_amostra, random_state=42)

# Plotar dados originais
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(dados['idade'], dados['renda'], alpha=0.5, color='blue')
plt.title('Dados Originais')
plt.xlabel('Idade')
plt.ylabel('Renda')

# Plotar dados da amostra aleatória simples
plt.subplot(1, 2, 2)
plt.scatter(amostra_aleatoria_simples['idade'], amostra_aleatoria_simples['renda'], alpha=0.5, color='green')
plt.title('Amostra Aleatória Simples')
plt.xlabel('Idade')
plt.ylabel('Renda')

plt.tight_layout()
plt.show()
```

# Viés de amostragem

```python
# Uma amostra que não representa a população
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados aleatórios
np.random.seed(42)
idade = np.random.randint(20, 60, 100)
renda = idade * np.random.randint(200, 500) + np.random.normal(0, 10000, 100)

# Aplicar viés de seleção (selecionar apenas pessoas com mais de 30 anos)
idade_selecionada = idade[idade > 30]
renda_selecionada = renda[idade > 30]

# Plotar dados originais
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(idade, renda, alpha=0.5, color='blue')
plt.title('Dados Originais')
plt.xlabel('Idade')
plt.ylabel('Renda')

# Plotar dados com viés de seleção
plt.subplot(1, 2, 2)
plt.scatter(idade_selecionada, renda_selecionada, alpha=0.5, color='red')
plt.title('Dados com Viés de Seleção')
plt.xlabel('Idade')
plt.ylabel('Renda')

plt.tight_layout()
plt.show()
```

# Média Amostral versus Média Populacional

## Média Amostral

```python
# É o cálculo da média considerando apenas os elementos de uma amostra. É como calcular a nota média de apenas uma turma daquela mesma escola.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gerar dados aleatórios
np.random.seed(42)
idade = np.random.randint(20, 60, 100)
renda = idade * np.random.randint(200, 500) + np.random.normal(0, 10000, 100)

# Criar DataFrame
dados = pd.DataFrame({'idade': idade, 'renda': renda})

# Amostragem aleatória simples
tamanho_amostra = 30
amostra_aleatoria_simples = dados.sample(n=tamanho_amostra, random_state=42)

# Calcular a média amostral
media_amostral_idade = amostra_aleatoria_simples['idade'].mean()
media_amostral_renda = amostra_aleatoria_simples['renda'].mean()

print(f"Média amostral da idade: {media_amostral_idade}")
print(f"Média amostral da renda: {media_amostral_renda}")

# Plotar dados originais
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(dados['idade'], dados['renda'], alpha=0.5, color='blue')
plt.axhline(y=media_amostral_renda, color='red', linestyle='--', label='Média Amostral da Renda')
plt.axvline(x=media_amostral_idade, color='red', linestyle='--', label='Média Amostral da Idade')
plt.title('Dados Originais')
plt.xlabel('Idade')
plt.ylabel('Renda')
plt.legend()

# Plotar dados da amostra aleatória simples
plt.subplot(1, 2, 2)
plt.scatter(amostra_aleatoria_simples['idade'], amostra_aleatoria_simples['renda'], alpha=0.5, color='green')
plt.axhline(y=media_amostral_renda, color='red', linestyle='--', label='Média Amostral da Renda')
plt.axvline(x=media_amostral_idade, color='red', linestyle='--', label='Média Amostral da Idade')
plt.title('Amostra Aleatória Simples')
plt.xlabel('Idade')
plt.ylabel('Renda')
plt.legend()

plt.tight_layout()
plt.show()
```

## Média Populacional

```python
# É o cálculo da média considerando todos os elementos da população. É como calcular a nota média de todos os alunos de uma escola.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gerar dados aleatórios
np.random.seed(42)
idade = np.random.randint(20, 60, 100)
renda = idade * np.random.randint(200, 500) + np.random.normal(0, 10000, 100)

# Criar DataFrame
dados = pd.DataFrame({'idade': idade, 'renda': renda})

# Calcular a média populacional
media_populacional_idade = dados['idade'].mean()
media_populacional_renda = dados['renda'].mean()

print(f"Média populacional da idade: {media_populacional_idade}")
print(f"Média populacional da renda: {media_populacional_renda}")

# Plotar dados originais com média populacional
plt.figure(figsize=(12, 6))

plt.scatter(dados['idade'], dados['renda'], alpha=0.5, color='blue')
plt.axhline(y=media_populacional_renda, color='red', linestyle='--', label='Média Populacional da Renda')
plt.axvline(x=media_populacional_idade, color='red', linestyle='--', label='Média Populacional da Idade')
plt.title('Dados Originais com Média Populacional')
plt.xlabel('Idade')
plt.ylabel('Renda')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Gerar dados aleatórios
np.random.seed(42)
idade = np.random.randint(20, 60, 100)
renda = idade * np.random.randint(200, 500) + np.random.normal(0, 10000, 100)

# Criar DataFrame
dados = pd.DataFrame({'idade': idade, 'renda': renda})

# Amostragem aleatória simples
tamanho_amostra = 30
amostra_aleatoria_simples = dados.sample(n=tamanho_amostra, random_state=42)

# Calcular a média populacional
media_populacional_idade = dados['idade'].mean()
media_populacional_renda = dados['renda'].mean()

# Calcular a média amostral
media_amostral_idade = amostra_aleatoria_simples['idade'].mean()
media_amostral_renda = amostra_aleatoria_simples['renda'].mean()

# Criar tabela de comparação
tabela_comparacao = pd.DataFrame({
    'Média': ['Idade', 'Renda'],
    'Populacional': [media_populacional_idade, media_populacional_renda],
    'Amostral': [media_amostral_idade, media_amostral_renda]
})

print(tabela_comparacao)
```
