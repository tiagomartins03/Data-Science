# BoxPlot

```python
# É um diagrama gráfico que representa a distribuição dos dados
```

- Mínimo (Minimum): O valor mais baixo no conjunto de dados, excluindo outliers.
- Primeiro quartil (Q1): O valor abaixo do qual 25% dos dados estão situados.
- Mediana (Q2): O valor central que divide o conjunto de dados em duas metades iguais.
- Terceiro quartil (Q3): O valor abaixo do qual 75% dos dados estão situados.
- Máximo (Maximum): O valor mais alto no conjunto de dados, excluindo outliers.

- Caixa (Box): Representa o intervalo interquartil (IQR), que é a distância entre o primeiro e o terceiro quartil (Q3 - Q1). A mediana é marcada dentro da caixa.
- Bigodes (Whiskers): Extendem-se dos limites da caixa até os valores mínimos e máximos dentro de 1.5 vezes o IQR a partir dos quartis (Q1 - 1.5IQR até Q3 + 1.5IQR).
- Outliers: Valores fora do alcance dos bigodes são considerados outliers e são geralmente representados como pontos individuais.

## Outliers

Para encontrar outliers em um conjunto de dados usando um boxplot, utiliza-se a regra do intervalo interquartil (IQR). A fórmula para determinar os outliers é a seguinte:

1. **Calcular o IQR (Intervalo Interquartil)**:
\[ \text{IQR} = Q3 - Q1 \]

2. **Definir os limites para os outliers**:
   - Limite inferior:
   \[ \text{Limite Inferior} = Q1 - 1.5 \times \text{IQR} \]
   - Limite superior:
   \[ \text{Limite Superior} = Q3 + 1.5 \times \text{IQR} \]

3. **Identificar os outliers**:
   - Qualquer valor menor que o limite inferior é considerado um outlier.
   - Qualquer valor maior que o limite superior é considerado um outlier.

Portanto, os outliers são definidos como:

\[ \text{Outlier Inferior} < Q1 - 1.5 \times \text{IQR} \]
\[ \text{Outlier Superior} > Q3 + 1.5 \times \text{IQR} \]

Onde:
- \( Q1 \) é o primeiro quartil (25º percentil).
- \( Q3 \) é o terceiro quartil (75º percentil).

### Exemplo:

Se tivermos um conjunto de dados com \( Q1 = 10 \) e \( Q3 = 20 \):

1. Calcular o IQR:
\[ \text{IQR} = 20 - 10 = 10 \]

2. Determinar os limites para os outliers:
\[ \text{Limite Inferior} = 10 - 1.5 \times 10 = 10 - 15 = -5 \]
\[ \text{Limite Superior} = 20 + 1.5 \times 10 = 20 + 15 = 35 \]

3. Identificar os outliers:
   - Valores menores que -5 são considerados outliers inferiores.
   - Valores maiores que 35 são considerados outliers superiores.

Qualquer valor no conjunto de dados que esteja fora desse intervalo \([-5, 35]\) será considerado um outlier.

### Exemplo

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Definir intervalo de valores
min_val = 10
max_val = 50

# Gerar alguns dados de exemplo dentro do intervalo [min_val, max_val]
np.random.seed(10)
data = np.random.uniform(min_val, max_val, 100)

# Imprimir os números randômicos
print("Números randômicos:")
print(data)

# Calcular os valores de Q1, Q2 (mediana), Q3, mínimo e máximo
q1 = np.percentile(data, 25)
median = np.percentile(data, 50)
q3 = np.percentile(data, 75)
min_value = np.min(data[data >= q1 - 1.5 * (q3 - q1)])
max_value = np.max(data[data <= q3 + 1.5 * (q3 - q1)])

# Imprimir os valores de Q1, Q2, Q3, mínimo e máximo
print(f"\nQ1: {q1:.2f}")
print(f"Mediana (Q2): {median:.2f}")
print(f"Q3: {q3:.2f}")
print(f"Min: {min_value:.2f}")
print(f"Max: {max_value:.2f}")

# Criar um boxplot
plt.figure(figsize=(8, 6))
box = sns.boxplot(data=data)

# Adicionar rótulos
plt.text(0, min_value, f'Min: {min_value:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, q1, f'Q1: {q1:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, median, f'Median: {median:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, q3, f'Q3: {q3:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, max_value, f'Max: {max_value:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')

# Adicionar títulos e rótulos
plt.title('Boxplot de Dados de Exemplo')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Percentil

```python
# O percentil é uma medida estatística que indica a posição relativa de um valor em um conjunto de dados ordenados. 
# Especificamente, o p-ésimo percentil é o valor abaixo do qual p% dos dados se encontram. Por exemplo, se um aluno está no 90º percentil em um exame, isso significa que ele teve uma pontuação maior do que 90% dos outros alunos que fizeram o exame.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Dados para a distribuição normal
mean = 0
std_dev = 1
x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x, mean, std_dev)

# Percentis e seus respectivos valores Z
percentis = [0.13, 2.28, 15.8, 50, 84.2, 97.72, 99.87]
z_scores = stats.norm.ppf([p/100 for p in percentis])

# Configuração do gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribuição Normal')

# Marcação dos percentis
for p, z in zip(percentis, z_scores):
    plt.axvline(z, color='red', linestyle='--')
    plt.text(z, stats.norm.pdf(z, mean, std_dev), f'p {p}', color='red', ha='center')

# Adiciona rótulos e título
plt.xlabel('Escore Z')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição Normal com Percentis Marcados')
plt.legend()

# Mostra o gráfico
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados aleatórios
np.random.seed(0)
dados = np.random.normal(loc=50, scale=15, size=1000)

# Calcular percentis
percentis = [25, 50, 75]  # Percentis desejados
percentis_valores = np.percentile(dados, percentis)

# Calcular a CDF dos dados
sorted_dados = np.sort(dados)
cdf = np.arange(1, len(sorted_dados) + 1) / len(sorted_dados)

# Plotar a CDF
plt.figure(figsize=(10, 6))
plt.plot(sorted_dados, cdf, marker='.', linestyle='none', label='CDF')

# Adicionar linhas para os percentis
for perc, valor in zip(percentis, percentis_valores):
    plt.axvline(valor, linestyle='--', color='red', label=f'{perc}º Percentil ({valor:.2f})')

# Configuração do gráfico
plt.title('Gráfico de Percentis')
plt.xlabel('Valor')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

# Tabela de Frequência

```python
# Uma tabela de frequências de uma variável divide a amplitude variável em segmentos espaçados e nos diz quantos valores caem em cada segmento.
```

```python
import pandas as pd

# Suponha que temos um conjunto de dados de idades de um grupo de pessoas
dados = [23, 25, 23, 30, 25, 25, 28, 30, 23, 28, 35, 30, 28, 23, 25, 28, 35, 40, 30, 28]

# Criação de um DataFrame com os dados
df = pd.DataFrame(dados, columns=['Idade'])

# Criação da tabela de frequência
tabela_frequencia = df['Idade'].value_counts().reset_index()
tabela_frequencia.columns = ['Idade', 'Frequência']
tabela_frequencia = tabela_frequencia.sort_values(by='Idade').reset_index(drop=True)

# Exibição da tabela de frequência
print(tabela_frequencia)
```

# Histograma

```python
# É a representação gráfico dos resultaos obtidos na tabela de frequência
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que temos um conjunto de dados de idades de um grupo de pessoas
dados = [23, 25, 23, 30, 25, 25, 28, 30, 23, 28, 35, 30, 28, 23, 25, 28, 35, 40, 30, 28]

# Criação de um DataFrame com os dados
df = pd.DataFrame(dados, columns=['Idade'])

# Criação da tabela de frequência
tabela_frequencia = df['Idade'].value_counts().reset_index()
tabela_frequencia.columns = ['Idade', 'Frequência']
tabela_frequencia = tabela_frequencia.sort_values(by='Idade').reset_index(drop=True)

# Exibição da tabela de frequência
print(tabela_frequencia)

# Criação do histograma
plt.figure(figsize=(10, 6))
plt.hist(df['Idade'], bins=range(20, 45, 5), edgecolor='black')

# Configuração do gráfico
plt.title('Histograma das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.xticks(range(20, 45, 5))

# Mostra o histograma
plt.show()
```

## Função de densidade (KDE)

```python
# O gráfico de densidade possui a mesma função do histograma, mas representa a distribuição como uma curva contínua. 
# A altura da curva no eixo vertical é uma estimativa da probabilidade de um elemento possuir aquele valor.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Suponha que temos um conjunto de dados de idades de um grupo de pessoas
dados = [23, 25, 23, 30, 25, 25, 28, 30, 23, 28, 35, 30, 28, 23, 25, 28, 35, 40, 30, 28]

# Criação de um DataFrame com os dados
df = pd.DataFrame(dados, columns=['Idade'])

# Criação da tabela de frequência
tabela_frequencia = df['Idade'].value_counts().reset_index()
tabela_frequencia.columns = ['Idade', 'Frequência']
tabela_frequencia = tabela_frequencia.sort_values(by='Idade').reset_index(drop=True)

# Exibição da tabela de frequência
print(tabela_frequencia)

# Criação do histograma e função de densidade usando seaborn
plt.figure(figsize=(10, 6))
sns.histplot(df['Idade'], bins=range(20, 45, 5), kde=True, edgecolor='black')

# Configuração do gráfico
plt.title('Histograma e Função de Densidade das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Suponha que temos um conjunto de dados de idades de um grupo de pessoas
dados = [23, 25, 23, 30, 25, 25, 28, 30, 23, 28, 35, 30, 28, 23, 25, 28, 35, 40, 30, 28]

# Criação de um DataFrame com os dados
df = pd.DataFrame(dados, columns=['Idade'])

# Criação da tabela de frequência
tabela_frequencia = df['Idade'].value_counts().reset_index()
tabela_frequencia.columns = ['Idade', 'Frequência']
tabela_frequencia = tabela_frequencia.sort_values(by='Idade').reset_index(drop=True)

# Exibição da tabela de frequência
print(tabela_frequencia)

# Criação do gráfico de densidade usando seaborn
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Idade'], fill=True, color='blue', bw_adjust=0.5)

# Configuração do gráfico
plt.title('Gráfico de Densidade das Idades')
plt.xlabel('Idade')
plt.ylabel('Densidade')

# Mostrar o gráfico
plt.show()
```
