# BoxPlot

```python
# É um diagrama gráfico que representa a distribuição dos dados
```

- Mínimo (Minimum): O valor mais baixo no conjunto de dados, excluindo outliers.
- Primeiro quartil (Q1): O valor abaixo do qual 25% dos dados estão situados.
- Mediana (Q2): O valor central que divide o conjunto de dados em duas metades iguais.
- Terceiro quartil (Q3): O valor abaixo do qual 75% dos dados estão situados.
- Máximo (Maximum): O valor mais alto no conjunto de dados, excluindo outliers.

- Caixa (Box): Representa o intervalo interquartil (IQR), que é a distância entre o primeiro e o terceiro quartil (Q3 - Q1). A mediana é marcada dentro da caixa.
- Bigodes (Whiskers): Extendem-se dos limites da caixa até os valores mínimos e máximos dentro de 1.5 vezes o IQR a partir dos quartis (Q1 - 1.5IQR até Q3 + 1.5IQR).
- Outliers: Valores fora do alcance dos bigodes são considerados outliers e são geralmente representados como pontos individuais.

## Outliers

Para encontrar outliers em um conjunto de dados usando um boxplot, utiliza-se a regra do intervalo interquartil (IQR). A fórmula para determinar os outliers é a seguinte:

1. **Calcular o IQR (Intervalo Interquartil)**:
\[ \text{IQR} = Q3 - Q1 \]

2. **Definir os limites para os outliers**:
   - Limite inferior:
   \[ \text{Limite Inferior} = Q1 - 1.5 \times \text{IQR} \]
   - Limite superior:
   \[ \text{Limite Superior} = Q3 + 1.5 \times \text{IQR} \]

3. **Identificar os outliers**:
   - Qualquer valor menor que o limite inferior é considerado um outlier.
   - Qualquer valor maior que o limite superior é considerado um outlier.

Portanto, os outliers são definidos como:

\[ \text{Outlier Inferior} < Q1 - 1.5 \times \text{IQR} \]
\[ \text{Outlier Superior} > Q3 + 1.5 \times \text{IQR} \]

Onde:
- \( Q1 \) é o primeiro quartil (25º percentil).
- \( Q3 \) é o terceiro quartil (75º percentil).

### Exemplo:

Se tivermos um conjunto de dados com \( Q1 = 10 \) e \( Q3 = 20 \):

1. Calcular o IQR:
\[ \text{IQR} = 20 - 10 = 10 \]

2. Determinar os limites para os outliers:
\[ \text{Limite Inferior} = 10 - 1.5 \times 10 = 10 - 15 = -5 \]
\[ \text{Limite Superior} = 20 + 1.5 \times 10 = 20 + 15 = 35 \]

3. Identificar os outliers:
   - Valores menores que -5 são considerados outliers inferiores.
   - Valores maiores que 35 são considerados outliers superiores.

Qualquer valor no conjunto de dados que esteja fora desse intervalo \([-5, 35]\) será considerado um outlier.

### Exemplo

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Definir intervalo de valores
min_val = 10
max_val = 50

# Gerar alguns dados de exemplo dentro do intervalo [min_val, max_val]
np.random.seed(10)
data = np.random.uniform(min_val, max_val, 100)

# Imprimir os números randômicos
print("Números randômicos:")
print(data)

# Calcular os valores de Q1, Q2 (mediana), Q3, mínimo e máximo
q1 = np.percentile(data, 25)
median = np.percentile(data, 50)
q3 = np.percentile(data, 75)
min_value = np.min(data[data >= q1 - 1.5 * (q3 - q1)])
max_value = np.max(data[data <= q3 + 1.5 * (q3 - q1)])

# Imprimir os valores de Q1, Q2, Q3, mínimo e máximo
print(f"\nQ1: {q1:.2f}")
print(f"Mediana (Q2): {median:.2f}")
print(f"Q3: {q3:.2f}")
print(f"Min: {min_value:.2f}")
print(f"Max: {max_value:.2f}")

# Criar um boxplot
plt.figure(figsize=(8, 6))
box = sns.boxplot(data=data)

# Adicionar rótulos
plt.text(0, min_value, f'Min: {min_value:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, q1, f'Q1: {q1:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, median, f'Median: {median:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, q3, f'Q3: {q3:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')
plt.text(0, max_value, f'Max: {max_value:.2f}', horizontalalignment='center', verticalalignment='bottom', color='blue')

# Adicionar títulos e rótulos
plt.title('Boxplot de Dados de Exemplo')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```
