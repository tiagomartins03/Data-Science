# Regressão à Média

```python
# É a tendência de que valores extremos voltem a média ao longo do tempo
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados sintéticos
np.random.seed(0)
x = np.random.normal(loc=50, scale=10, size=100)  # Dados com média 50 e desvio padrão 10

# Introduzir valores extremos
x[::10] = np.random.normal(loc=100, scale=5, size=10)  # A cada 10 valores, adicionar um valor extremo

# Calcular a média antes e depois
mean_before = np.mean(x)
mean_after = np.mean(np.random.choice(x, size=100, replace=True))

# Plotar os dados
plt.figure(figsize=(12, 6))
plt.plot(x, 'o', label='Dados Originais')
plt.axhline(mean_before, color='r', linestyle='--', label=f'Média Antes: {mean_before:.2f}')
plt.axhline(mean_after, color='g', linestyle='--', label=f'Média Depois: {mean_after:.2f}')
plt.title('Regressão à Média')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.legend()
plt.show()
```
