A regressão linear simples modela o relacionamento entre a magnitude de uma variável e aquela de uma segunda - por exemplo, conforme X aumenta, Y aumenta ou vice-versa. A correlação é outro jeito de medir como duas variáveis estão relacionadas. A diferença é que enquanto a correlação mede a força de uma associação entre duas variáveis, a regressão quantifica a natureza do relacionamento.

## Regressão Linear Simples

A regressão linear simples modela o relacionamento entre a magnitude de uma variável e a de uma segunda — por exemplo, conforme \( X \) aumenta, \( Y \) aumenta ou vice-versa. A correlação também mede como duas variáveis estão relacionadas. A diferença é que, enquanto a correlação mede a força de uma associação entre duas variáveis, a regressão quantifica a natureza desse relacionamento.

## A Equação de Regressão

A regressão linear simples estima exatamente quanto \( Y \) mudará quando \( X \) variar por uma certa quantidade. Com a correlação, as variáveis \( X \) e \( Y \) são intercambiáveis. Com a regressão, buscamos prever \( Y \) a partir de \( X \) usando um relacionamento linear.

A equação da regressão linear simples é dada por:

$$Y = \beta_0 + \beta_1X + \epsilon$$

Onde:

- **\( Y \)**: Variável dependente  
- **\( \beta_0 \)**: Intercepto  
- **\( \beta_1 \)**: Coeficiente angular  
- **\( X \)**: Variável independente  
- **\( \epsilon \)**: Erro aleatório  

O símbolo **\( \beta_0 \)** é conhecido como intercepto (ou constante), e o símbolo **\( \beta_1 \)** é o coeficiente angular de **\( X \)**. Ambos aparecem nos resultados do R como coeficientes, embora o termo "coeficiente" seja comumente reservado para **\( \beta_1 \)**. A variável **\( Y \)** é chamada de _resposta_ ou _variável dependente_, pois depende de **\( X \)**, que é conhecida como _preditora_ ou _variável independente_.




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Gerando dados aleatórios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # Variável independente
y = 4 + 3 * X + np.random.randn(100, 1)  # Variável dependente com um pouco de ruído

# Criando o modelo de regressão
model = LinearRegression()
model.fit(X, y)

# Prevendo valores usando o modelo
X_new = np.array([[0], [2]])  # Valores de X para prever Y
y_predict = model.predict(X_new)

# Coeficientes da regressão
beta_0 = model.intercept_[0]
beta_1 = model.coef_[0][0]

# Gráfico antes da regressão
plt.figure(figsize=(14, 6))

# Gráfico original
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Dados Observados')
plt.title('Gráfico Antes da Regressão')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Gráfico após a regressão
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', label='Dados Observados')
plt.plot(X_new, y_predict, color='red', label='Linha de Regressão')

# Adicionando a linha do intercepto
plt.axhline(y=beta_0, color='green', linestyle='--', label='Intercepto (β₀)')

# Adicionando a equação da reta no gráfico
equation_text = f'Y = {beta_0:.2f} + {beta_1:.2f}X'
plt.text(0.5, 15, equation_text, fontsize=12, color='red', ha='center')  # Ajuste a coordenada Y aqui

# Ajustando os limites do eixo Y
plt.ylim(0, 25)  # Definindo um limite superior menor para reduzir o espaço

plt.title('Gráfico Após a Regressão')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.tight_layout()
plt.show()
```
