# Distribuição de Weibull


A distribuição de Weibull é uma distribuição de probabilidade contínua, amplamente utilizada em análises de confiabilidade, estudo de tempos de vida de produtos e na modelagem de tempos até a falha de sistemas mecânicos. Ela foi introduzida pelo engenheiro sueco Waloddi Weibull em 1951 e é conhecida por sua flexibilidade em modelar diferentes tipos de dados de vida útil.

### Características Principais:

1. **Parâmetro de Forma (k ou β)**:
   - Esse parâmetro determina a forma da distribuição.
   - Para \( k < 1 \): A distribuição é decrescente e indica que a taxa de falha diminui com o tempo, como em sistemas onde as falhas ocorrem mais no início da vida útil.
   - Para \( k = 1 \): A distribuição se torna uma distribuição exponencial, o que significa que a taxa de falha é constante ao longo do tempo.
   - Para \( k > 1 \): A distribuição tem uma curva em forma de sino, indicando que a taxa de falha aumenta com o tempo, comum em produtos que sofrem desgaste e falham mais frequentemente com o uso prolongado.

2. **Parâmetro de Escala (λ ou η)**:
   - Este parâmetro afeta a escala da distribuição e é semelhante ao conceito de vida média ou característica da vida útil.
   - Um valor maior de \( \lambda \) desloca a distribuição para a direita, indicando tempos de vida mais longos.

3. **Parâmetro de Localização (μ)** (opcional):
   - Esse parâmetro, se incluído, desloca a distribuição ao longo do eixo dos tempos de vida.

### Função Densidade de Probabilidade (PDF):

A função densidade de probabilidade da distribuição de Weibull é dada por:

$$
f(x; k, \lambda) = \begin{cases} 
\frac{k}{\lambda} \left(\frac{x}{\lambda}\right)^{k-1} e^{-\left(\frac{x}{\lambda}\right)^k} & \text{se } x \geq 0, \\
0 & \text{se } x < 0.
\end{cases}
$$

Onde:
- \( k \) é o parâmetro de forma,
- \( \lambda \) é o parâmetro de escala,
- \( x \) é o tempo de vida ou a variável de interesse.

### Aplicações:

- **Análise de Confiabilidade**: Para prever o tempo de falha de componentes e sistemas, o que é vital na engenharia e manufatura.
- **Engenharia de Manutenção**: Para planejar manutenções preventivas com base em tempos estimados de falha.
- **Economia e Finanças**: Para modelar o tempo até a ocorrência de certos eventos, como a falência de uma empresa.

Devido à sua flexibilidade, a distribuição de Weibull pode ser ajustada a uma ampla gama de dados empíricos, tornando-se uma ferramenta essencial em diversas disciplinas.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Definindo os parâmetros da distribuição de Weibull
shape_param = 1.5  # Parâmetro de forma (k)
scale_param = 2000  # Parâmetro de escala (lambda), vida média em horas

# Gerando um conjunto de dados usando a distribuição de Weibull
data = weibull_min.rvs(shape_param, scale=scale_param, size=1000)

# Criando os bins para o histograma
bins = np.linspace(0, np.max(data), 50)

# Calculando a probabilidade para cada bin usando a PDF da distribuição de Weibull
pdf_values = weibull_min.pdf(bins, shape_param, scale=scale_param)

# Normalizando para que a soma das probabilidades seja igual a 1
bin_width = bins[1] - bins[0]
pdf_values_normalized = pdf_values * bin_width

# Plotando o histograma com a função densidade de probabilidade
plt.figure(figsize=(10, 6))
plt.hist(data, bins=bins, density=True, alpha=0.6, color='blue', label='Histograma')
plt.plot(bins, pdf_values_normalized, 'r-', lw=2, label='PDF da Weibull')

# Adicionando detalhes ao gráfico
plt.title("Distribuição de Weibull - Tempo de Falha de Componentes", fontsize=15)
plt.xlabel("Tempo de Falha (horas)", fontsize=12)
plt.ylabel("Probabilidade", fontsize=12)
plt.legend()

# Mostrando o gráfico
plt.show()
```
