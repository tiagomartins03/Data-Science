# Distribuição de Poisson

A distribuição de Poisson, nomeada em homenagem ao matemático francês Siméon Denis Poisson, é uma distribuição de probabilidade que pode ser utilizada para descrever o número de ocorrências de um evento em um determinado intervalo de tempo ou espaço, quando temos as seguintes características:

- Utilizada quando os eventos são raros e independentes
- Utilizada quando a taxa média de ocorrências é constante
- Utilizada quando os resultados são contagens (números inteiros não negativos)

## Princípios

- Os eventos são raros e independentes;
- A taxa média de ocorrências é constante;
- Os resultados são contagens (números inteiros não negativos);

## Condições

- Os eventos devem ser independentes entre si, ou seja, a ocorrência de um evento não deve afetar a probabilidade de ocorrência de outro.
- A taxa média de ocorrências deve ser constante ao longo do intervalo analisado, garantindo que a distribuição de Poisson seja aplicável a processos com taxas consistentes.
- As ocorrências devem ser raras em relação ao intervalo analisado, o que significa que a chance de ocorrência em um intervalo específico é baixa.
- O que está sendo modelado é uma contagem de ocorrências, o que torna a distribuição de Poisson adequada para processos de contagem de eventos raro

## Exemplo

Uma loja recebe em médias 5 ligações a cada 10 minutos. Qual a probabilidade de receber 8 por dia? 

$P(X = x) = \frac{e^{-\lambda} \lambda^x}{x!}$

$P(X = 8) = \frac{e^{-5} 5^8}{8!}$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo o número médio de chamadas por minuto (lambda)
lambda_calls = 10

# Gerando um conjunto de dados usando a distribuição de Poisson
data = np.random.poisson(lambda_calls, 1000)

# Plotando o histograma com Seaborn para uma melhor visualização
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=range(0, max(data)+1), kde=False, color="blue")

# Adicionando detalhes ao gráfico
plt.title("Distribuição de Poisson - Chamadas por Minuto em um Call Center", fontsize=15)
plt.xlabel("Número de Chamadas por Minuto", fontsize=12)
plt.ylabel("Frequência", fontsize=12)
plt.xticks(range(0, max(data)+1))

# Mostrando o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Definindo o número médio de chamadas por minuto (lambda)
lambda_calls = 10

# Criando um range de valores possíveis para o número de chamadas
x = np.arange(0, 30)

# Calculando a probabilidade de cada valor de x usando a PMF da distribuição de Poisson
pmf = poisson.pmf(x, lambda_calls)

# Plotando a função de massa de probabilidade
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='blue')

# Adicionando detalhes ao gráfico
plt.title("Distribuição de Poisson - Probabilidade de Chamadas por Minuto", fontsize=15)
plt.xlabel("Número de Chamadas por Minuto", fontsize=12)
plt.ylabel("Probabilidade", fontsize=12)
plt.xticks(x)

# Mostrando o gráfico
plt.show()
```
