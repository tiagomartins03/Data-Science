# Distribuição Binomial

A distribuição binomial é uma distribuição de probabilidade discreta que descreve o número de ‘sucessos’ (ou ‘falhas’) em uma série de experimentos independentes e idênticos. Ela é utilizada para calcular a probabilidade de um determinado número de sucessos/falhas, dado o número total de tentativas e a probabilidade de sucesso/falha em cada tentativa.

## Aplicações

## Formula

A fórmula para calcular a probabilidade de obter exatamente ‘x’ sucessos em ‘n’ tentativas é dada por: P(X = x) = (n x) p^x (1 – p)^(n-x).

A fórmula é composta por três partes: (n x), p^x e (1 – p)^(n-x).
Onde ‘n’ é o número total de tentativas, ‘x’ é o número de sucessos desejado e ‘p’ é a probabilidade de sucesso em cada tentativa.
Essa fórmula é uma ferramenta estatística que permite calcular a probabilidade de um determinado número de sucessos em um número fixo de tentativas independentes, com uma taxa de sucesso constante. A fórmula é dada por P(X = x) = (nCx) * (p^x) * ((1 – p)^(n-x)), onde n é o número total de tentativas, x é o número de sucessos desejado, p é a probabilidade de sucesso em uma única tentativa e (1 – p) é a probabilidade de falha em uma única tentativa.

A distribuição binomial é amplamente utilizada em situações onde há um número fixo de tentativas independentes, cada uma com a mesma probabilidade de sucesso.
A fórmula da distribuição binomial é essencial para calcular a probabilidade de ocorrência de um número específico de sucessos em um conjunto de tentativas.
A compreensão da fórmula da distribuição binomial é fundamental para análises estatísticas em diversas áreas, como engenharia, medicina, economia e controle de qualidade.

## Aplicação

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parâmetros da distribuição binomial
n = 10  # Número de experimentos (número de componentes selecionados)
p = 0.05  # Probabilidade de sucesso (encontrar um componente defeituoso)

# Calculando a probabilidade de encontrar exatamente 2 componentes defeituosos
k = 2
probability_k_defective = stats.binom.pmf(k, n, p)
print(f'Probabilidade de encontrar exatamente {k} componentes defeituosos: {probability_k_defective:.4f}')

# Gerando a distribuição binomial para todos os possíveis valores de k
x = np.arange(0, n+1)
binom_pmf = stats.binom.pmf(x, n, p)

# Plotando a distribuição binomial
plt.figure(figsize=(10, 6))
plt.bar(x, binom_pmf, color='skyblue')
plt.plot(k, probability_k_defective, 'ro', label=f'P(X={k}) = {probability_k_defective:.4f}', markersize=12)

# Personalizando o gráfico
plt.title('Distribuição Binomial - Controle de Qualidade')
plt.xlabel('Número de Componentes Defeituosos (k)')
plt.ylabel('Probabilidade')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Exibindo o gráfico
plt.show()
```
